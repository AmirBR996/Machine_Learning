### Imports
This cell imports the necessary libraries for data manipulation (`pandas`), numerical operations (`numpy`), and plotting (`matplotlib.pyplot`). These are foundational libraries for most data science tasks in Python.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Load Dataset
This cell loads the Diabetes dataset from `sklearn.datasets`. `X` will contain the feature variables (independent variables), and `y` will contain the target variable (dependent variable, which is a measure of disease progression).

```python
from sklearn.datasets import load_diabetes
X , y = load_diabetes(return_X_y=True)
```

### Initialize Linear Regression Model
Here, a `LinearRegression` model is initialized from `sklearn.linear_model`. This model will be used to fit a linear relationship between the features (`X`) and the target variable (`y`).

```python
from sklearn.linear_model import LinearRegression
linearregression = LinearRegression()
```

### Train and Predict with Linear Regression
This cell trains the `LinearRegression` model using the `fit` method, providing the feature matrix `X` and the target vector `y`. After training, it uses the `predict` method to generate predictions (`y_pred_linear`) based on the `X` features.

```python
linearregression.fit(X ,y)
y_pred_linear = linearregression.predict(X)
```

### Evaluate Linear Regression
This cell imports the `r2_score` metric from `sklearn.metrics` to evaluate the performance of the `LinearRegression` model. The R-squared score measures the proportion of variance in the dependent variable that can be predicted from the independent variables.

```python
from sklearn.metrics import r2_score
print(f"r2_score using linear_regression : {r2_score(y , y_pred_linear)}")
```

### Initialize, Train, and Predict with Ridge Regression
This cell introduces `Ridge` regression, a type of linear regression that includes L2 regularization to prevent overfitting. A `Ridge` model is initialized with a small `alpha` (regularization strength), then trained on the data, and finally used to make predictions (`y_pred_ridge`).

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=0.0001)
ridge.fit(X, y)

y_pred_ridge = ridge.predict(X)
```

### Evaluate Ridge Regression
Similar to the linear regression evaluation, this cell calculates and prints the R-squared score for the `Ridge` regression model, providing an assessment of its predictive performance.

```python
print(f"r2_score using ridge_regression : {r2_score(y , y_pred_ridge)}")
```
