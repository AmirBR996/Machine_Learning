This SGD class implements linear regression using stochastic gradient descent. It initializes the weights (coef_) and bias (intercept_) to zero, then trains the model in the fit() method by looping over a specified number of epochs. In each epoch, it shuffles the dataset and updates the weights and bias per sample using the gradient of the mean squared error with respect to each parameter, scaled by the learning rate (lr). The predict() method computes predictions for new inputs by applying the learned weights and bias. This approach allows the model to iteratively learn optimal coefficients even on large datasets by updating after each individual sample rather than the whole dataset.

```python
import numpy as np

class SGD():
    def __init__(self, lr=0.01, epoch=1000):
        self.coef_ = None
        self.intercept_ = None
        self.lr = lr
        self.epoch = epoch

    def fit(self, X, y):
        n_samples, n_features = X.shape

        self.coef_ = np.zeros(n_features)
        self.intercept_ = 0

        for _ in range(self.epoch):
            indices = np.arange(n_samples)
            np.random.shuffle(indices)

            for i in indices:
                y_hat = self.intercept_ + np.dot(X[i], self.coef_)
                error = y[i] - y_hat
                db = -2 * error
                dw = -2 * X[i] * error
                self.intercept_ -= self.lr * db
                self.coef_ -= self.lr * dw

    def predict(self, X):
        return self.intercept_ + np.dot(X, self.coef_)
```

```python
from sklearn.datasets import make_regression

bgd = SGD(lr=0.5, epoch=10)

bgd.fit(X, y)

print("Intercept:", bgd.intercept_)
print("Slope:", bgd.coef_)
```

```python
y_pred_batch = bgd.predict(X)
```

r2_score: This tells you how well your model explains the variance. A score of 0.51 means the model captures about 51% of the pattern in the data.

```python
r2_score(y , y_pred_batch)
```

### Suggestions & Best Practices
1.  **Train-Test Split**: Currently, you are training and testing on the *same* data. In a real project, always use `train_test_split` to see how your model performs on data it hasn't seen before.
2.  **Learning Rate Sensitivity**: In your custom `BatchGD`, if the `lr` is too high, the model might 'explode' (get infinite errors); if it's too low, it will be very slow. Always experiment with values like 0.1, 0.01, and 0.001.
3.  **Vectorization**: Your custom class correctly uses `np.dot`, which is much faster than using Python `for` loops for math. This is a great practice.
4.  **Convergence**: Instead of a fixed number of epochs, you could add a 'tolerance' to stop training early if the error stops decreasing.
