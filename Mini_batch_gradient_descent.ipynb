Mini-batch Gradient Descent is an optimization algorithm used to train machine learning models, especially in neural networks. It is a compromise between Batch Gradient Descent and Stochastic Gradient Descent (SGD).

```python
import numpy as np
import random

class MiniBGD:
    def __init__(self, lr=0.01, epochs=1000, batch_size=10):
        self.coef_ = None
        self.intercept_ = None
        self.lr = lr
        self.epochs = epochs
        self.batch_size = batch_size

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.coef_ = np.zeros(n_features)
        self.intercept_ = 0

        for epoch in range(self.epochs):
            # Shuffle the data at the start of each epoch
            indices = np.arange(n_samples)
            np.random.shuffle(indices)
            X_shuffled = X[indices]
            y_shuffled = y[indices]

            # Process mini-batches
            for start in range(0, n_samples, self.batch_size):
                end = start + self.batch_size
                X_batch = X_shuffled[start:end]
                y_batch = y_shuffled[start:end]

                # Prediction
                y_hat = np.dot(X_batch, self.coef_) + self.intercept_

                # Compute gradients
                error = y_batch - y_hat
                dw = -2 * np.dot(X_batch.T, error) / X_batch.shape[0]
                db = -2 * np.sum(error) / X_batch.shape[0]

                # Update parameters
                self.coef_ -= self.lr * dw
                self.intercept_ -= self.lr * db

    def predict(self, X):
        return np.dot(X, self.coef_) + self.intercept_
```

```python
from sklearn.datasets import make_regression

minibgd = MiniBGD(lr=0.5, epochs=200, batch_size = 10)

minibgd.fit(X, y)

print("Intercept:", minibgd.intercept_)
print("Slope:", minibgd.coef_)
```

```python
y_pred_batch = minibgd.predict(X)
```

r2_score: This tells you how well your model explains the variance. A score of 0.51 means the model captures about 51% of the pattern in the data.

```python
r2_score(y , y_pred_batch)
```

### Suggestions & Best Practices
1.  **Train-Test Split**: Currently, you are training and testing on the *same* data. In a real project, always use `train_test_split` to see how your model performs on data it hasn't seen before.
2.  **Learning Rate Sensitivity**: In your custom `BatchGD`, if the `lr` is too high, the model might 'explode' (get infinite errors); if it's too low, it will be very slow. Always experiment with values like 0.1, 0.01, and 0.001.
3.  **Vectorization**: Your custom class correctly uses `np.dot`, which is much faster than using Python `for` loops for math. This is a great practice.
4.  **Convergence**: Instead of a fixed number of epochs, you could add a 'tolerance' to stop training early if the error stops decreasing.
